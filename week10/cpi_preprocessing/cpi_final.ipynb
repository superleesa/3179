```python
import pandas as pd
```

```python
df = pd.read_csv("cpi_raw.csv")
```

```python
df
```

```python
df = df[(df["Unit"] == "Index") & (df["Measure"] == "Index")]
```

```python
len(df["Subject"].unique())
```

```python
df
```

```python
df[["Subject", "Country", "Time", "Value"]].groupby(["Country", "Time"]).size()
```

```python
df = df.loc[df["Country"] != "Japan"]
```

```python
df = df[["Subject", "Country", "Time", "Value"]]
```

```python
df.columns = ["Subject", "Country", "Year", "CPI"]
```

```python
df2 = pd.read_csv("cpi_japan_raw.csv")
```

```python
df2
```

```python
df["Country"].unique()
```

```python
df2 = df2[(df2["Unit"] == "Index") & (df2["Measure"] == "Index") & (df2["Frequency"] == "Annual")]
```

```python
df2 = df2[["Subject", "Country", "Time", "Value"]]
```

```python
df2.columns = ["Subject", "Country", "Year", "CPI"]
```

```python
df_oup = pd.concat([df, df2], ignore_index=False)
```

```python
df_oup = df_oup.loc[df_oup["Country"] != "Türkiye"]
```

```python
df_oup.loc[df_oup["Country"] == "China (People's Republic of)", "Country"] = ["China" for _ in range((df_oup["Country"] == "China (People's Republic of)").sum())]
```

```python
df_oup.to_csv("cpi.csv", index=False)
```
