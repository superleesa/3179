```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("raw_avg_wage.csv")
```

```python
df
```

```python
all(df[["Country", "Time", "Value"]].groupby(["Country", "Time"]).size() == 1)
```

```python
df_oup = df[["Country", "Time", "Value"]]
```

```python
df_oup.to_csv("avg_wage.csv", index=False)
```

```python
countries = df_oup.groupby("Country")
for country_name, country_df in countries:
    plt.plot(country_df["Time"], country_df["Value"], label=country_name, color="red" if country_name == "Japan" else "grey")
```
